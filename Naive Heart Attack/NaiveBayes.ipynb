```python
import pandas as pd
import numpy as np
```

Membaca dataset

```python
dataset = pd.read_csv("heart.csv")

dataset.info()
```

Membagi data independent(X) 0-12 dan Dependent(y) 13

```python
X = dataset.iloc[:,:13]
y = dataset.iloc[:,13]
```

Melakukan split data (data train dan data test)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

membuat model

```python
from sklearn.naive_bayes import GaussianNB
Naive_Bayes = GaussianNB()
Naive_Bayes.fit(X_train, y_train.ravel())
```

```python
y_predik = Naive_Bayes.predict(X_test)
y_predik
```

cek akurasi

```python
from sklearn.metrics import accuracy_score, classification_report
akurasi = accuracy_score(y_test, y_predik)
print("Nilai Akurasi = ", akurasi)
print(classification_report(y_test,y_predik))
```

melakukan tes dengan data baru

```python
data_baru = pd.read_excel("databaru_test_heart.xlsx")
x_baru = data_baru.iloc[:,:]
y_predikbaru = Naive_Bayes.predict(x_baru)
y_predikbaru

data_baru['Heart Attack'] = y_predikbaru
data_baru
```
