Read Dataset

```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("waterQuality1.csv")
```

```python
dataset.info()
```

```python
dataset['ammonia'] = dataset['ammonia'].replace('#NUM!', float(0))
dataset['ammonia'] = dataset['ammonia'].astype(float)
dataset['is_safe'] = dataset['is_safe'].replace('#NUM!', int(0))
dataset['is_safe'] = dataset['is_safe'].astype(int)
```

Terdapat beberapa data pada kolom ammonia dan is_safe yang memiliki nilai #NUM sehingga terjadi eror. Disini #NUM diganti menjadi 0 dan merubah typedata dari ammonia menjadi float dan is safe menjadi int

```python
dataset.info()
```

Split Variable

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,20].values
```

Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

Split data train $ test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

Klasifikasi KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

Training / Fiting Model

```python
knn = KNeighborsClassifier(n_neighbors =27 ,metric='minkowski').fit(x_train,y_train)
knn
```

Prediksi

```python
a = np.array([[1000,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100,100]])
y_prediksi = knn.predict(x_test)
y_prediksi
```

Cek Akurasi

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print("Nilai Akurasi :", accuracy_score(y_test,y_prediksi))
```

Mencari nilai K terbaik

```python
n_akurasi =[]
for k in range(1,300):
    k_terbaik = KNeighborsClassifier(n_neighbors=k,metric='minkowski').fit(x_train,y_train)
    y_prediksi = k_terbaik.predict(x_test)
    akurasi = accuracy_score(y_test,y_prediksi)
    n_akurasi.append(akurasi)

print(n_akurasi)
K_terbesar, akurasi_terbesar = max(enumerate(n_akurasi), key=lambda x: x[1])

print("K = ", K_terbesar + 1)
print("Nilai Nilai Akurasi = ", akurasi_terbesar)
```
